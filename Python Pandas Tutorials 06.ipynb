```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sample_data_tutorial_06.csv')
df
```

```python
newdf = df.replace(-99999,np.NaN)
newdf
```

```python
newdf = df.replace([-99999, -88888],np.NaN)
newdf
```

```python
newdf = df.replace({
        'temperature': -99999,
        'windspeed': [-99999, -88888],
        'event': 'No event'
    }, np.NaN)
newdf
```

```python
# Podemos gerar um mapa das alterações que queremos fazer:
newdf = df.replace({
    -99999: np.NaN,
    -88888: np.NaN,
    'No event': 'Sunny'
})
newdf
```

```python
# Importando outro csv com algumas unidades que precisam ser limpas!
df = pd.read_csv('sample_data_tutorial_06a.csv')
df
```

```python
# É necessário usar o 'regex' (regular expression)
# No caso abaixo estamos substituindo todas as letras (de A a Z - maiúscula e minúscula) por vazio (='')
newdf = df.replace('[A-Za-z]', '', regex=True)
newdf
```

```python
# Observe, no caso anterior, que ele removeu o que pedimos mas também removeu toda coluna 'event'
# Para fazer as substituições somente em determinadas colunas é preciso utilizar o dicionário:
newdf = df.replace({
    'temperature': '[A-Za-z]', 
    'windspeed': '[A-Za-z]'
}, '', regex=True)
newdf
```
