```python
import pandas as pd
```

```python
df = pd.read_csv('Engine_deck_example.csv')
df
```

```python
galt = df.groupby(['Altitude'])
galt
```

```python
gmach = df.groupby(['Mach'])
gmach
```

```python
# O comando anterior agrupará as Altitude, DISA e Mach como "keys" e o resto como valores
for alt, val_df in galt:
    print(alt)
    print (val_df)
```
